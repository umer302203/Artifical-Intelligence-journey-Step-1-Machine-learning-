So first we will import all libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Now we will import data on which we want to train our model

```python
df = pd.read_csv("C:/Users/Umer/Documents/Code Workplace/Pyton/Ai_Project/Datasets/creditcard.csv")
```

```python
df.sample(71)
```

Now as we know that time is not important so we will ignore and then class is ur target feature

```python
X =df.drop(["Time","Class"],axis =1)
Y = df["Class"]
```

Now we will take overview of whole data

```python
df.describe()
```

Now convert the all data into numpy arrays for sklearn

```python
X_np =X.to_numpy()
Y_np =Y.to_numpy()
```

Now we will convert data into standard form 

```python
from sklearn import preprocessing
Std = preprocessing.StandardScaler()
Standard_pc = Std.fit_transform(X_np)
```

Now we will split data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Standard_pc,Y_np,test_size=0.2,random_state =42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
original_model = model.fit(X_train,y_train)
```

Now we will predict on test data

```python
Prediction = original_model.predict(X_test)
Prediction[:10]
```

Now we will check the probility

```python
Probility = original_model.predict_proba(X_test)
Probility[:11]
```

Let evalute the model

```python
from sklearn.metrics import log_loss
log_loss(y_test,Probility)
```
